```python
#1.create a numpy array with values from 10 to 49 (inclusive).
#● Reshape it into a 4x10 matrix.extract the last column from the reshaped matrix
import numpy as np
a1 = np.arange(10,50)
print(a1)
print(a1.shape)
a12 = np.reshape(a1,(4,10))
print("the reshaped array\n",a12)
print("The last column\n",a12[:,-1:])
```

```python
#2.Create two 3×3 matrices: One filled with all 5s , One with values from 1 to 9
#● Add the two matrices ● Multiply them element-wise ● Then find the transpose of the result
a21 = np.array([[5,5,5],[5,5,5],[5,5,5]])
a22 = np.arange(1,10)
a23 = np.reshape(a22,(3,3))
print("array one\n",a21)
print("array 2\n",a23)
sum = np.add(a21, a23)
print("Added value\n",sum)
print("Product\n",a21*a23)
print("Transpose is\n",(a21*a23).T)
```

```python
#3.Create a NumPy array of 15 random integers between 10 and 100.
#● Replace all values greater than 50 with -1 ● Count how many values were replaced.
a3 = np.random.randint(10,100,size=(5,3))
a31 = []
count=0
print("The array is\n",a3)
flat_a3 = a3.flatten()
for i in range(15):
  if flat_a3[i]>50:
    a31.append(-1)
    count+=1
  else:
    a31.append(flat_a3[i])
a31 = np.reshape(a31,(5, 3))
print("replaced\n",a31)
print("count is ",count)
```

```python
#4.Generate a 5×4 array with random values between 0 and 1 (float).
#● Find the row with the maximum sum of values. ● Normalize each row using the formula:x-min/(max-min)
a4 = np.random.random((5, 4))
print("The array is\n", a4)
sum4 = np.sum(a4, axis=1)
print("Sum of rows:", sum4)
row_index = np.argmax(sum4)
print("The row with maximum sum is:", a4[row_index])
min_vals = a4.min(axis=1).reshape(-1,1)  # min of each row as column vector
max_vals = a4.max(axis=1).reshape(-1,1)  # max of each row as column vector

a4_norm = (a4 - min_vals)/(max_vals-min_vals)
print("Normalized array:\n",a4_norm)
```

```python
#5. arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
#● Find all unique elements ● Find the duplicates (elements that appear more than once)
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique_elements, counts = np.unique(arr, return_counts=True)
unique = unique_elements[counts==1]
print("All unique elements:",unique)
duplicates = unique_elements[counts>1]
print("Duplicate elements:",duplicates)
```

```python
#Create an array: a = np.array([[1, 2], [3, 4]])
#● Append a new row [5, 6] ● Then append a new column [7, 8, 9]
a = np.array([[1, 2], [3, 4]])
row = np.array([[5, 6]])
result = np.append(a,row,axis=0)
print("row appended\n",result)
column = np.array([[7],[8],[9]])
result1 = np.append(result,column,axis=1)
print("column appended\n",result1)
```

```python
#a = np.array([1, 2, 3]) b = np.array([4, 5, 6]) c = np.array([7, 8, 9])
#● Join them into a single array ● Reshape the result into a 3×3 matrix ● Extract the diagonal elements
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
s=np.concatenate((a,b,c))
print("single array:",s)
s1=np.reshape(s,(3,3))
print("reshaped\n",s1)
diag = np.diagonal(s1)
print("Main diagonal:", diag)
```

```python
#Create a NumPy array of integers from 1 to 20.
#● Extract all values that are divisible by 3 or 5. ● Count how many such values exist.
a9 = np.random.randint(1,21,size=(4,5))
a91 = []
count1=0
print("The array is\n",a9)
flat_a9 = a9.flatten()
for i in flat_a9:
  if i%3==0 or i%5==0:
    a91.append(int(i))
    count1+=1
print("the numbers:\n",a91)
print("count is ",count1)
```

Pandas activity

```python
#1.Load the Titanic dataset and display the first 10 rows.
import pandas as pd
import seaborn as sns
data = sns.load_dataset('titanic')
df=pd.DataFrame(data)
print(df.head(10))
```

```python
#2. Find the number of missing values in each column.
df.isnull().sum()
```

```python
#3. Find the percentage of passengers who survived, grouped by gender.
df.groupby('sex')['survived'].mean()*100
```

```python
#4. Find the average, median, and standard deviation of passenger ages.
print("Average",df["age"].mean())
print("Median",df["age"].median())
print("Standard deviation",df["age"].std())
```

```python
#5.Count how many passengers were in each class (pclass).
df['pclass'].value_counts()
```

```python
#6. Show all passengers who were female, under age 18, and survived.
df[(df['sex']<="female") & (df['age']<=18) & (df['survived']==1)]
```

```python
#7. Replace missing age values with the median age.
print(df['age'].fillna(df['age'].median()))
print(df.isnull().sum())
```

```python
#8.Create a column age_group with values:
#a. 'Child' for age < 18 b. 'Adult' for age 18–60 c. 'Senior' for age > 60 d. Use: apply() or np.where() or custom function
def classify_age(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(classify_age)
print(df.head())
```

```python
#9.Display the top 5 passengers who paid the highest fare.
top_fare = df.sort_values(by='fare',ascending=False).head(5)
print(top_fare)
```

```python
#10. Find the average fare paid, grouped by both pclass and sex.
df.groupby(['sex','pclass'])['fare'].mean()
```

```python
#11.Rename the following columns:
#a. 'sex' → 'gender' b. 'fare' → 'ticket_fare'
df=df.rename(columns={'sex': 'gender', 'fare': 'ticket_fare'})
print(df.head())
```

```python
#12.Count the number of passengers who survived and didn’t survive, grouped by class.
df.groupby('class')['survived'].value_counts()
```

```python
#13. Find the average age of passengers grouped by both class and gender.

df.groupby(['class','gender'])['age'].mean()
```
