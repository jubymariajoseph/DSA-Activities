STANDADIZING DATA

```python
renamed_data1 = data1.rename(columns={'name':'country'})
renamed_data3 = data3.rename(columns={'Country':'country'})
```

```python
print(data.info())#Convert numeric columns to appropriate types, Handle missing or erroneous data
print(renamed_data1.info())
print(data2.info())
print(renamed_data3.info())
```

There is no missing values in each of the dataset and columns are numeric in nature

```python
print(data3)
```

```python
merged_data = pd.merge(data, renamed_data1, on='country', how='inner')
merged_data = pd.merge(merged_data, data2, on='country', how='inner')
merged_data = pd.merge(merged_data, renamed_data3, on='country', how='inner')
```

Merged dataset

```python
merged_data
```

```python
merged_data.info()
```

Create new columns such as internet penetration rate (internet users / population × 100)

```python
merged_data['internet penetration rate'] = (merged_data['internet_users']/merged_data['population'])*100
```

```python
merged_data
```

```python
print(merged_data.isnull().sum()) # Validate the data for consistency and completeness
print(merged_data[merged_data.isnull().any(axis=1)])
print(merged_data.duplicated().sum())
```

Validated the data for consistency and completeness. Duplicates or missing data isn't found in the dataset

```python
#Find countries with highest internet penetration rates

rate = merged_data['internet penetration rate'].max()
country = merged_data[merged_data['internet penetration rate']==rate]
print(country['country'])
```

```python
# Calculate average literacy rate across countries

literacy = merged_data['literacy_rate'].mean()
print("Average is ",literacy)
```

```python
correlation_matrix = merged_data[['GDP', 'population', 'internet_users', 'literacy_rate']].corr()
print(correlation_matrix)
```

```python
import sqlite3
conn = sqlite3.connect("country.db")
cursor = conn.cursor()
```

```python
merged_data.to_sql("Final_database",conn,if_exists = "replace",index=True)
```

```python
cursor.execute('select country from Final_database order by "internet penetration rate" desc limit 1')
```

```python
print(cursor.fetchone()) # Countries with highest internet penetration
```

```python
#  Average literacy rate
cursor.execute("select avg(literacy_rate) from Final_database")
print("Average literacy rate is ")
print(cursor.fetchone())
```

```python
#  Correlation between literacy rate and internet penetration

cursor.execute('select (count(*)*sum(literacy_rate * "internet penetration rate")- sum(literacy_rate)*sum("internet penetration rate"))/(sqrt(count(*)*sum(literacy_rate * literacy_rate)- sum(literacy_rate)*sum(literacy_rate))*sqrt(count(*)*sum("internet penetration rate"*"internet penetration rate")- sum("internet penetration rate")*sum("internet penetration rate"))) from Final_database')
print("correlation is ")
print(cursor.fetchone())
```

```python
#  GDP per capita analysis

cursor.execute('select country, (GDP/population) as GDP_per_capita from Final_database')
print(cursor.fetchall())
```

```python
#   Which countries have a GDP per capita above $10,000?

cursor.execute('select country,(GDP/population) as GDP_per_capita from Final_database where GDP_per_capita > 10000')
print("The countries are")
print(cursor.fetchall())
```

```python
#What is the total population covered in the dataset?

cursor.execute('select sum(population) from Final_database')
print("The total polpulation is")
print(cursor.fetchone())
```

```python
#Which countries have the lowest literacy rates, and how does that impact internet access

cursor.execute('select country, literacy_rate, "internet penetration rate" from Final_database order by literacy_rate limit 1')
print("The countries are")
print(cursor.fetchall())
```

As the literacy rate gets lower with reduced use of internet.

```python
#What are the top 5 wealthiest countries by total GDP, and how does that compare with population size?

cursor.execute('select country, GDP, population,(GDP/population) as GDP_per_capita from Final_database order by GDP desc limit 5')
print("The countries are")
print(cursor.fetchall())
```

when population is increasing GDP is decreasing and GDP_per_capita is decreasing. As population increases the earning of per person decreases.

```python
#Find countries where internet users exceed 70% of the population.

cursor.execute('select country,internet_users, (internet_users*100/population) as percentage from Final_database where percentage >70')
print("The countries are")
print(cursor.fetchall())
```

```python
#What is the average GDP per capita for countries with internet penetration above 50%

cursor.execute('select  avg(GDP/population) from Final_database where "internet penetration rate" >50')
print("The average is")
print(cursor.fetchone())
```

```python
#How many countries have a literacy rate above 90%, and what is their average internet penetration

cursor.execute('select count(*), avg("internet penetration rate") from Final_database where literacy_rate>90')
print("The average and count are")
print(cursor.fetchone())
```
