```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].median())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].median())
```

Group by Director: avg Rating, Revenue

```python
director_stats = df.groupby('Director')[['Rating', 'Revenue (Millions)']].mean().reset_index()
director_stats.rename(columns={'Rating': 'Avg_Rating', 'Revenue (Millions)': 'Avg_Revenue'}, inplace=True)

print(director_stats)
```

Group by Year: count, avg Rating

```python
year_stats = df.groupby('Year').agg(Movie_count=('Movie_Title','count'),Avg_Ratings = ('Rating','mean')).reset_index()
year_stats
```

Create a new column Profitability:
a) Revenue divided by Duration
b) Find top 10 most profitable movies

```python
df['Profitability'] = df['Revenue (Millions)']/df['Runtime (Minutes)']
df_prof = df.sort_values(by='Profitability',ascending=False)
df_prof.head(10)
```

Create a column that categorizes movies by rating:
a) Rating < 6: Poor
b) 6 ≤ Rating < 7.5: Average
c) Rating ≥ 7.5: Excellent

```python
conditions = [df['Rating'] < 6, (df['Rating'] >= 6) & (df['Rating'] < 7.5)]

choices = ['Poor', 'Average']

df['category'] = np.select(conditions, choices, default='Excellent')
df
```

Create a custom function to score movies based on:
1) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
2) Apply a function to dataset to assign the score

```python
def high_score(row):
  return row['Rating'] * 10 + row['Revenue (Millions)']
df['High score'] = df.apply(high_score, axis=1)
df
```
